```python
import tensorflow as tf
from tensorflow import keras
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
```

```python
(X_train, y_train), (X_test, y_test)= keras.datasets.mnist.load_data()
```

```python
len(X_train)
```

```python
len(X_test)
```

```python
X_train[0].shape
```

```python
X_train[0]
```

This mnist data set includes 60, 000 images in the train set and 10, 000 images as test set. When we see the 1st image size it's 28 x 28 .It's a 2D vector. That vector entries includes values ranging from 0 to 255.

```python
plt.matshow(X_train[0])
```

Using matplotlib's matshow we can visualize the digit includes in the first image in the training set. It seems to be 5.

```python
y_train[0]
```

The y values represent the integer values which that image represents. That y values range from 0 to 9.

```python
y_train[:5]
```

This array includes first 5 records in the y_train set.

Now we convert this 2D array into a 1D vector. For that use reshape.

```python
X_train.shape
```

60, 000 of images, each 28 x 28 2D array

```python
X_train_flattened= X_train.reshape(len(X_train), 28*28)
X_train_flattened.shape
```

Now each image has converted to a 1D vector. Do same for test set as well.

```python
X_test_flattened= X_test.reshape(len(X_test), 28*28)
X_test_flattened[0]
```

Now let's create a simple neural network which includes input layer and output layer. Later we add hidden layers.

Now 784 neurons are in the input layer, output layer has 10 neurons as output is among 10 elements (0 - 9).

```python
model = keras.Sequential([keras.layers.Dense(10, input_shape=(784,), activation='sigmoid')])
model.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy']
)
model.fit(X_train_flattened, y_train, epochs=5)
```

```python
model.evaluate(X_test_flattened,y_test)
```

```python
model.predict(X_test_flattened)
```

```python
plt.matshow(X_test[0])
```

```python
y_predicted= model.predict(X_test_flattened)
y_predicted[0]
```

This is a 10 value array. It represents the score value for each element in output layer. Output should be the element which got a higher score value. Let's run the argmax function on numpy for that.

```python
np.argmax(y_predicted[0])
```

Wrongly predicted. Actual is 7 but we got 3, Later we add hidden layers and try to improve the accuracy of the model.

Now let's visualize a confusion matrix

```python
y_predicted_labels= [np.argmax(i) for i in y_predicted]
cm = tf.math.confusion_matrix(labels= y_test, predictions= y_predicted_labels)

import seaborn as sns
plt.figure(figsize=(10,7))
sns.heatmap(cm, annot=True, fmt= 'd')
plt.xlabel('predicted')
plt.ylabel('Truth')
```

X_train and X_test values should be scaled. Then we can increase the accuracy. X_train values range from 0 to 255. So divide the set by 255.

```python
X_train= X_train/ 255
X_test= X_test/ 255
```

Let's run the same code which was written for simple neural network model and see the accuracy.

```python
X_train_flattened= X_train.reshape(len(X_train), 28*28)
X_test_flattened= X_test.reshape(len(X_test), 28*28)

model = keras.Sequential([keras.layers.Dense(10, input_shape=(784,), activation='sigmoid')])
model.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy']
)
model.fit(X_train_flattened, y_train, epochs=5)

model.evaluate(X_test_flattened,y_test)

y_predicted= model.predict(X_test_flattened)

y_predicted_labels= [np.argmax(i) for i in y_predicted]
cm = tf.math.confusion_matrix(labels= y_test, predictions= y_predicted_labels)

import seaborn as sns
plt.figure(figsize=(10,7))
sns.heatmap(cm, annot=True, fmt= 'd')
plt.xlabel('predicted')
plt.ylabel('Truth')
```

Now you can see the accuracy has increased. In heatmap the non dagonal entries have also been reduced, cuz now the wrongly predicted value count has been decreased.

```python
np.argmax(y_predicted[0])
```

Earlier the 1st record in X_test was predicted as 3 wrongly, now it has been correctly predicted as 7.

Let's add a hidden layer and see the accuracy. This time we use 100 as the hidden layer number of neurons, it's a guess. and we'll use Relu as the activation function. Then last layer no need to mention the input_shape, it'll be identified. Also the diagonal entry value has been increased, means count of correctly predicted have increased.

```python
model = keras.Sequential([keras.layers.Dense(100, input_shape=(784,), activation='relu'),
                          keras.layers.Dense(10, activation='sigmoid')
                          ])
model.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy']
)
model.fit(X_train_flattened, y_train, epochs=5)

model.evaluate(X_test_flattened,y_test)

y_predicted= model.predict(X_test_flattened)

y_predicted_labels= [np.argmax(i) for i in y_predicted]
cm = tf.math.confusion_matrix(labels= y_test, predictions= y_predicted_labels)

import seaborn as sns
plt.figure(figsize=(10,7))
sns.heatmap(cm, annot=True, fmt= 'd')
plt.xlabel('predicted')
plt.ylabel('Truth')
```

Observation: Accuracy has been increased.In heatmap the non dagonal entries have also been reduced, cuz now the wrongly predicted value count has been decreased. Also the diagonal entry value has been increased, means count of correctly predicted have increased.

Now , here we add 2 hidden layers, let's see the accuracy.

```python
model = keras.Sequential([keras.layers.Dense(100, input_shape=(784,), activation='relu'),
                          keras.layers.Dense(50, activation='relu'),
                          keras.layers.Dense(10, activation='sigmoid')
                          ])
model.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy']
)
model.fit(X_train_flattened, y_train, epochs=5)

model.evaluate(X_test_flattened,y_test)

y_predicted= model.predict(X_test_flattened)

y_predicted_labels= [np.argmax(i) for i in y_predicted]
cm = tf.math.confusion_matrix(labels= y_test, predictions= y_predicted_labels)

import seaborn as sns
plt.figure(figsize=(10,7))
sns.heatmap(cm, annot=True, fmt= 'd')
plt.xlabel('predicted')
plt.ylabel('Truth')
```

Instead of using reshape, there's a function called 'Flatten' in keras. Let's use it here. No need to use X_train_flattened and X_test_flattened. Btw no change in accuracy here. Just the same code except a wording.

```python
model = keras.Sequential([keras.layers.Flatten(input_shape=(28, 28)),
                          keras.layers.Dense(100, activation='relu'),
                          keras.layers.Dense(10, activation='sigmoid')
                          ])
model.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy']
)
model.fit(X_train, y_train, epochs=5)

model.evaluate(X_test,y_test)

y_predicted= model.predict(X_test)

y_predicted_labels= [np.argmax(i) for i in y_predicted]
cm = tf.math.confusion_matrix(labels= y_test, predictions= y_predicted_labels)

import seaborn as sns
plt.figure(figsize=(10,7))
sns.heatmap(cm, annot=True, fmt= 'd')
plt.xlabel('predicted')
plt.ylabel('Truth')
```
